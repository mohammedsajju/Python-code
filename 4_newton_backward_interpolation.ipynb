```python
import numpy as np

def newton_backward_interpolation(x, y, x_target):
    n = len(x)
    h = x[1] - x[0]

    # Create backward difference table
    diff = np.zeros((n, n))

    # Fill first column with y values
    for i in range(n):
        diff[i][0] = y[i]

    # Compute backward differences
    for j in range(1, n):
        for i in range(n-1, j-1, -1):
            diff[i][j] = diff[i][j-1] - diff[i-1][j-1]

    # Calculate u
    u = (x_target - x[n-1]) / h

    # Initialize result
    y_target = diff[n-1][0]
    u_product = 1
    factorial = 1

    # Apply formula
    for i in range(1, n):
        u_product *= (u + i - 1)
        factorial *= i
        y_target += (u_product * diff[n-1][i]) / factorial

    return y_target


# Example data
x = np.array([1, 2, 3, 4])
y = np.array([1, 8, 27, 64])  # Example: y = x^3

# Value to find
x_target = 3.5

# Function call
result = newton_backward_interpolation(x, y, x_target)

print("Interpolated value at", x_target, "is", result)
```
