```python
# Simpson's 3/8 Rule Implementation

def f(x):
    return x**3 + 2*x   # Example function (change pannalaam)

# Input values
a = float(input("Enter lower limit: "))
b = float(input("Enter upper limit: "))
n = int(input("Enter number of subintervals: "))

# Check condition (n must be multiple of 3)
if n % 3 != 0:
    print("Number of subintervals must be a multiple of 3")
else:
    # Step size
    h = (b - a) / n

    # Initial sum
    s = f(a) + f(b)

    # Applying Simpson's 3/8 Rule
    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            s += 2 * f(x)
        else:
            s += 3 * f(x)

    # Final result
    result = (3 * h / 8) * s

    print("Approximate value of the integral =", result)
```
