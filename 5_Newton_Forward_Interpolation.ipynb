```python
import numpy as np

# Function to create forward difference table
def forward_difference_table(y, n):
    diff_table = [y.copy()]

    for i in range(1, n):
        col = []
        for j in range(n - i):
            value = diff_table[i-1][j+1] - diff_table[i-1][j]
            col.append(value)
        diff_table.append(col)

    return diff_table


# Newton Forward Interpolation function
def newton_forward(x, y, x_val):
    n = len(x)
    diff_table = forward_difference_table(y, n)

    h = x[1] - x[0]
    p = (x_val - x[0]) / h

    result = y[0]
    p_term = 1
    factorial = 1

    for i in range(1, n):
        p_term *= (p - (i - 1))
        factorial *= i
        result += (p_term * diff_table[i][0]) / factorial

    return result


# Example data
x = np.array([1, 2, 3, 4])
y = np.array([1, 8, 27, 64])  # Example: y = x^3

# Value to find
x_val = 2.5

# Function call
result = newton_forward(x, y, x_val)

print("Interpolated value at", x_val, "is", result)
```
