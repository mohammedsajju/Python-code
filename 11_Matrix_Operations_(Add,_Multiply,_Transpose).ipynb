```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

# Addition
print("Matrix Addition:")
for i in range(len(A)):
    for j in range(len(A[0])):
        print(A[i][j] + B[i][j], end=" ")
    print()

# Multiplication
print("Matrix Multiplication:")
for i in range(len(A)):
    for j in range(len(B[0])):
        total = 0
        for k in range(len(B)):
            total += A[i][k] * B[k][j]
        print(total, end=" ")
    print()

# Transpose
print("Transpose:")
for i in range(len(A[0])):
    for j in range(len(A)):
        print(A[j][i], end=" ")
    print()
```
