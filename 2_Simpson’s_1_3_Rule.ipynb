```python
# Simpson's 1/3 Rule Implementation

def f(x):
    return x**2   # Example function (change pannalaam)

# Input values
a = float(input("Enter lower limit (a): "))
b = float(input("Enter upper limit (b): "))
n = int(input("Enter number of subintervals (even number): "))

# Check if n is even
if n % 2 != 0:
    print("Error: n must be an even number.")
    exit()

# Step size
h = (b - a) / n

# Initial sum
result = f(a) + f(b)

# Applying Simpson's Rule
for i in range(1, n):
    x = a + i * h
    if i % 2 == 0:
        result += 2 * f(x)
    else:
        result += 4 * f(x)

# Final result
result = result * (h / 3)

print("Approximate value of the integral =", result)
```
