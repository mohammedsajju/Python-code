```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

# Define differential equation
def fun(x, y):
    return [y[1], -y[0]]

# Boundary conditions
def bc(ya, yb):
    return [ya[0], yb[0] - 1]

# Initial mesh
x = np.linspace(0, np.pi/2, 10)

# Initial guess
y = np.zeros((2, x.size))

# Solve BVP
sol = solve_bvp(fun, bc, x, y)

# Plot solution
x_plot = np.linspace(0, np.pi/2, 100)
y_plot = sol.sol(x_plot)[0]

plt.plot(x_plot, y_plot)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Solution using solve_bvp")
plt.show()
```
